```python
def reduce_mem_usage(df, verbose=True):
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    start_mem = df.memory_usage().sum() / 1024**2    
    for col in df.columns:
        col_type = df[col].dtypes
        if col_type in numerics:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            else:
                if c_min > np.finfo(np.float16).min and c_max < np.finfo(np.float16).max:
                    df[col] = df[col].astype(np.float16)
                elif c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)    
    end_mem = df.memory_usage().sum() / 1024**2
    if verbose: print('Mem. usage decreased to {:5.2f} Mb ({:.1f}% reduction)'.format(end_mem, 100 * (start_mem - end_mem) / start_mem))
    return df
```

Fitting a Neural Network to the above train data

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import cross_val_score
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
from keras.callbacks import ModelCheckpoint
from keras.models import load_model
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization, LeakyReLU
from keras.layers import Dropout
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint
import matplotlib.pyplot as plt
from keras.models import load_model
```

```python
import keras.backend as K

def custom_metric(y_true, y_pred):
    avg_loss = K.mean(K.abs(y_pred-y_true))
    return K.log(avg_loss)
```

```python
def model_(m):
    model = Sequential()

    model.add(Dense(50, input_dim=m))
    model.add(BatchNormalization())
    model.add(LeakyReLU(0.1))
    model.add(Dense(100))
    model.add(BatchNormalization())
    model.add(LeakyReLU(0.1))
    model.add(Dense(250))
    model.add(BatchNormalization())
    model.add(LeakyReLU(0.1))
    model.add(Dense(500))
    model.add(BatchNormalization())
    model.add(LeakyReLU(0.1))
    model.add(Dense(1000))
    model.add(BatchNormalization())
    model.add(LeakyReLU(0.1))
    model.add(Dense(500))
    model.add(BatchNormalization())
    model.add(LeakyReLU(0.1))
    model.add(Dense(250))
    model.add(BatchNormalization())
    model.add(LeakyReLU(0.1))
    
    model.add(Dense(100))
    model.add(BatchNormalization())
    model.add(LeakyReLU(0.1))
    model.add(Dense(50))
    model.add(BatchNormalization())
    model.add(LeakyReLU(0.1))
    
    model.add(Dense(1, activation='linear'))
    model.compile(loss='mse', optimizer='adam', metrics = [custom_metric])
    return model
```

```python
model = model_(df_scalar_train.shape[1])
```

```python
n_epochs = 10
batch_size = 2000
es = EarlyStopping(monitor='val_loss', verbose=1, patience=5)  # , mode='min'
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=10, min_lr=1e-6)
mc = ModelCheckpoint('best_model.h5', monitor='val_loss', verbose=0, save_best_only=True)  # , mode='min'
# fit model
history = model.fit(df_scalar_train, train['scalar_coupling_constant'], validation_split=0.2, epochs=n_epochs, shuffle=True,
                    batch_size = batch_size, verbose=0, callbacks=[es, reduce_lr, mc])
```

```python
best_model = load_model('best_model.h5')
```

```python
op_metrics = best_model.evaluate(df_scalar_train, train['scalar_coupling_constant'])

op_metrics
```

## Metrics Interpretation & Further Improvements

The regressed parameters given by the model are a good fit compared to a simple model whose log loss would be very high (1 or above).

However, looking at the competition results where the Leaderboard touched log-loss in the range of [-3, -3.5], there is certainly a room for improvement such as:

1. In our model, we only used Mulliken charges fitted on test. Similar model fitting of the dipole moment data can be done so that we can even use it as a parameter.
2. Improving the model architecture can be done. Presently, we are only creating an MLP network. However, using custom network (using PyTorch) can help in achieving the required precision in our Scalar Coupling predictions.
