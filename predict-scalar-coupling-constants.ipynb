# III. Creating a model to fit the 4 Scalar coupling data features

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
df_train = pd.read_pickle(f'../input/molecular-properties-data-create/df_train_new.pkl')
```

```python
df_scalar = pd.read_csv(f'../input/champs-scalar-coupling/scalar_coupling_contributions.csv')
```

Fitting on the data, we merge the data with the required scalar_coupling data

```python
df_train = df_train.sample(frac = 0.5)
df_train.head()
```

```python
df_train = df_train.merge(df_scalar, on = ['molecule_name', 'atom_index_0', 'atom_index_1'])
```

```python
df_train.drop(columns= ['id', 'molecule_name','atom_index_0', 'atom_index_1', 'type_y', 'scalar_coupling_constant'], inplace=True)
```

```python
df_train.head()
```

Now that all the columns that aren't required for prediction are removed, we can fit the above cleaned features on the 4 factors, namely:
- fc
- sd
- pso
- dso

```python
target_y = ['fc', 'sd', 'pso', 'dso']

y_train = df_train[target_y]

X_train = df_train.drop(columns = target_y)
```

```python
X_train.shape
```

```python
X_train.head()
```

Encoding the multi-class labels of the molecule type

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(X_train['type_x'])
le.classes_
```

```python
X_train['type_x'] = le.transform(X_train['type_x']) 
```

```python
X_train.head()
```

## Data normalization

Since there are 4 fields of y, predicting them without normalization results in less accurate solutions (based on the last hidden layer) or would require separate model for each.
To circumvent this issue, we'll normalize the Outputs and restore the values after the predictions are made.

```python
y_mean = y_train.mean()
y_min = y_train.min()
y_max = y_train.max()
# y_min

'''
Note: To ensure all values lie within the (-1,1) range we make the following computation
'''
y_train = (y_train - y_mean) / (1.2 * (y_max - y_min))
y_train.head()
```

```python
X_train.head()
```

## Modeling using a Neural network

```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import cross_val_score
```

```python
N, m = X_train.shape
```

```python
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint
from keras.models import load_model
```

```python
# from sklearn.datasets import make_moons
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization, LeakyReLU
from keras.layers import Dropout
from keras.callbacks import EarlyStopping
from keras.callbacks import ModelCheckpoint
from matplotlib import pyplot
from keras.models import load_model
# generate 2d classification dataset
    
def train_model(X_train, y_train, n_epochs, model):
    # simple early stopping
    es = EarlyStopping(monitor='val_loss', verbose=1, patience=200)  # , mode='min'
    mc = ModelCheckpoint('best_model.h5', monitor='val_loss', verbose=0, save_best_only=True)  # , mode='min'
    # fit model
    history = model.fit(X_train, y_train, validation_split=0.2, epochs=n_epochs, verbose=0, callbacks=[es, mc])
    # load the saved model
    saved_model = load_model('best_model.h5')
    return saved_model
# evaluate the model
# saved_model = train_model(X_train[20_000:30_000], y_train[20_000:30_000], 1, None)
# _, train_acc = saved_model.evaluate(trainX, trainy, verbose=0)
# _, test_acc = saved_model.evaluate(testX, testy, verbose=0)
# print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
```

```python
index_lst = []

div = 5
diff = N//div

for j in range(div):
    index_lst.append((j)* diff)
    
index_lst.append(N) 
index_lst = np.array(index_lst)
```

```python
def iterative_train(X_train, y_train, indices, n_epochs= 10):
    model = model_()
    
    for i in range(len(indices)-1):  # 1
        print(indices[i+1])
        saved_model = train_model(X_train[indices[i]:indices[i+1]], y_train[indices[i]:indices[i+1]], n_epochs, model)
        model = saved_model
    return model
```

```python
# define model
def model_():
    model = Sequential()
    model.add(Dense(50, input_dim=m))
    model.add(Dropout(0.3))

    model.add(Dense(100))
    model.add(BatchNormalization())
    model.add(LeakyReLU(0.1))

    model.add(Dense(30))
    model.add(BatchNormalization())
    model.add(LeakyReLU(0.1))

    model.add(Dense(30))
    model.add(BatchNormalization())
    model.add(LeakyReLU(0.1))
    
    model.add(Dense(30))
    model.add(BatchNormalization())
    model.add(LeakyReLU(0.1))
   
    model.add(Dense(30))
    model.add(BatchNormalization())
    model.add(LeakyReLU(0.1))
    
    model.add(Dense(4, activation='tanh'))
    model.compile(loss='mae', optimizer='adam')
    return model
```

```python
model = iterative_train(X_train, y_train, index_lst) # index_lst
```
